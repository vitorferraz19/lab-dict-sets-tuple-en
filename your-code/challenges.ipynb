#### In the cell below, create a list named `sample_list_1` with 80 random values.

#### Convert `sample_list_1` to a set called `set1`. Print the length of the set. Is its length still 80?

#### Create another list named `sample_list_2` with 80 random values.

Requirements:

* Each value is an integer falling between 0 and 100.
* The values in the list don't have to be unique.

*Hint: Use a FOR loop.*

````python
sample_list_2 = []                         # Initializing from an empty list (sample_list_2) to start storing numbers.

for _ in range(80):                         # Setting the number of times the loop runs to 80 with no regard to the loop variable.
    sample_list_2.append(random.randint(0, 100))    # This random method will generate 80 numbers out of the 0 to 100 interval of integers with no necessary uniqueness and append these to the innitially empty list.

print(sample_list_2)
````
